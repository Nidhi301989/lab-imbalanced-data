```python
# Lab! Imbalance data

import pandas as pd
file_path = r"C:\Users\ashis\Downloads\data for ironheck\customer_churn.csv"
df = pd.read_csv(file_path)
print(df.head())
print(df.describe())
print(df.info())
```

```python
# Extract the target variable
y = df['Churn']
```

```python
from sklearn.preprocessing import StandardScaler

# Extract the independent variables
X = df[['tenure', 'SeniorCitizen', 'MonthlyCharges']]

# Scale the independent variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression

# Build the logistic regression model
model = LogisticRegression()
model.fit(X_scaled, y)
```

```python
from sklearn.metrics import accuracy_score

# Predict the target variable
y_pred = model.predict(X_scaled)

# Evaluate the model
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(X_scaled, y)

# Build and evaluate the logistic regression model
model_smote = LogisticRegression()
model_smote.fit(X_smote, y_smote)
y_smote_pred = model_smote.predict(X_smote)
accuracy_smote = accuracy_score(y_smote, y_smote_pred)
print(f'Accuracy after SMOTE: {accuracy_smote}')
```

```python
from imblearn.under_sampling import TomekLinks
from sklearn.model_selection import train_test_split

# Apply Tomek Links
tomek = TomekLinks()
X_tomek, y_tomek = tomek.fit_resample(X_scaled, y)

# Split the dataset after Tomek Links
X_train_tomek, X_test_tomek, y_train_tomek, y_test_tomek = train_test_split(X_tomek, y_tomek, test_size=0.3, random_state=42)

# Build and evaluate the logistic regression model
model_tomek = LogisticRegression()
model_tomek.fit(X_train_tomek, y_train_tomek)
y_tomek_pred = model_tomek.predict(X_test_tomek)
accuracy_tomek = accuracy_score(y_test_tomek, y_tomek_pred)
print(f'Accuracy after Tomek Links: {accuracy_tomek}')
```
